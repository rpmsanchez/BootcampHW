
## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
gender_group2 = purchase_data.groupby(["Gender"])
pur_count = pd.DataFrame(gender_group2["Purchase ID"].count())
avr_purprice = pd.DataFrame(gender_group2["Price"].mean())
tot_pv = pd.DataFrame(gender_group2["Price"].sum())
avr_tot_ppp = pd.DataFrame(tot_pv["Price"] / gen_count)

summary4 = pd.merge(pd.merge(pd.merge(pur_count, avr_purprice, on="Gender"), tot_pv, on="Gender"), avr_tot_ppp, on="Gender")
summary4 = summary4.rename(columns={"Purchase ID": "Purchase Count", "Price_x": "Average Purchase Price", "Price_y": "Total Purchase Value", 0 : "Average Total Purchase per Person"})
summary4
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 150]
age_labels = ["<10", "10-14", "15-20", "20-25", "25-30", "30-35", "35-40", "40+"]
unique["Age Group"] = pd.cut(unique["Age"], bins, labels=age_labels)
age_group = unique.groupby(["Age Group"])
age_count = pd.DataFrame(age_group["SN"].count())
age_sum = age_group["SN"].count().sum()
age_per = pd.DataFrame((age_count / age_sum) * 100)
age_per = age_per.rename(columns={"SN": "Percentage of Players"})

summary5 = pd.merge(age_count, age_per, on="Age Group")
summary5 = summary5.rename(columns={"SN": "Total Count"})
summary5
```
